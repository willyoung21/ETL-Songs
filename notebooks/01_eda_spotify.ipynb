```python
import pandas as pd

# Definir la ruta del archivo CSV
csv_path = r"C:\Users\USUARIO\Documents\Workshop2\data\raw\spotify_dataset.csv"

# Cargar el CSV en un DataFrame
df_spotify = pd.read_csv(csv_path)

# Configurar pandas para mostrar más filas y columnas
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

df_spotify.head(5)
```

```python
# Eliminar columnas innecesarias (suponiendo que algunas no son útiles para el análisis)
columns_to_drop = ['album_name', 'explicit', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 
                    'liveness', 'valence', 'time_signature']

df_cleaned_spotify = df_spotify.drop(columns=columns_to_drop)

# Verificar las primeras filas después de eliminar las columnas
print("Después de eliminar columnas innecesarias:")
# Configurar pandas para mostrar más filas y columnas
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

df_cleaned_spotify.head(5)
```

```python
# Revisar la cantidad de valores nulos por columna
print("\nValores nulos por columna:")
print(df_cleaned_spotify.isnull().sum())
```

```python
# Eliminar filas con valores nulos en las columnas 'artists' y 'track_name'
df_cleaned_spotify = df_cleaned_spotify.dropna(subset=['artists', 'track_name'])

# Verificar que ya no haya nulos
print(df_cleaned_spotify.isnull().sum())
```
